```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
plt.scatter(X, y);
plt.xlabel('Level');
plt.ylabel('Salary');
plt.title('Salary vs Level ($)')
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y);
```

```python
from sklearn.preprocessing import PolynomialFeatures
pol_reg = PolynomialFeatures(degree = 4)
X_pol = pol_reg.fit_transform(X)

lin_reg2 = LinearRegression()
lin_reg2.fit(X_pol, y);
```

```python
plt.figure(figsize = [13.5,4])
plt.subplot(1,2,1)
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X));
plt.title('Salary vs Level (Linear Regression)');
plt.xlabel('Level');
plt.ylabel('Salary');

plt.subplot(1,2,2)
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg2.predict(pol_reg.fit_transform(X)));
plt.title('Salary vs Level (Polynomial Regression)');
plt.xlabel('Level');
plt.ylabel('Salary');
```

```python
plt.figure(figsize = [13.5,4])
plt.subplot(1,2,1)
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X));
plt.title('Salary vs Level (Linear Regression)');
plt.xlabel('Level');
plt.ylabel('Salary');

plt.subplot(1,2,2)
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg2.predict(pol_reg.fit_transform(X_grid)));
plt.title('Salary vs Level (Polynomial Regression)');
plt.xlabel('Level');
plt.ylabel('Salary');
```
